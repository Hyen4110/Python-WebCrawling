```python
loc_num =0
df_total = pd.DataFrame(index=range(0,1), columns=['model','title1', 'title2','value'])


def get_data_function(urli):
    global loc_num
    global df_total
    url = urli
    headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
    req = requests.get(url,headers = headers)
    html = req.text
    soup = BeautifulSoup(html, 'lxml')


    #
    model_name = soup.find("h1", {"class":"specs-phone-name-title"}).text
    body = soup.find("div", {"id": "specs-list"})
    s =body.find_all("table",{"cellspacing":"0"})

    #s리스트 = 각 항목 섹션
    s_list  = []
    for i in s:
        s_list.append(i)

    #s[0]번 "네트워크"만 따로 추출
    key1 = s[0].find("tr",{"class":"tr-hover"})
    key_1 = key1.th.text
    key_2 = key1.find("td",{"class":"ttl"}).text
    val = key1.find("td",{"class":"nfo"}).text

    print(key_1)
    print(key_2)
    print(val)
    print("-------")

    
    df_total.loc[loc_num] = model_name, key_1, key_2, val
    loc_num+=1

    for i in range(1,len(s)):

        key1 = s[i]
        key_1 = key1.th.text

        ##2depth 항목명 가져오기
        key2 = key1.find_all("td",{"class":"ttl"})
        key2_list=[]
        for i in range(len(key2)):
            
            key2_list.append(key2[i].text)

        if "\xa0" in key2_list:
            key2_list.remove("\xa0")    

        ##끝에 있는 값 가져오기
        val_list = []    
        val = key1.find_all("td",{"class":"nfo"})
        for i in range(len(key2)):
            val_list.append(val[i].text.replace('"','').replace('\r\n','//').replace('\n','').replace('\r',''))

        if len(key2_list)<len(val_list):
            k = (len(val_list))-(len(key2_list)) 
            for i in range(k):
                key2_list.append("")

        key_1_list = []
        for i in range(len(key2_list)):
            key_1_list.append(key_1)

        model_list=[]
        for i in range(len(key2_list)):
            model_list.append(model_name)
        
            
        print(key_1_list)
        print(len(key2_list))
        print(len(val_list))
        print(key2_list)
        print(val_list)

        data1 = list(zip(model_list, key_1_list,key2_list, val_list))
        print(data1)

        for i in range(0,len(data1)):
            df_total.loc[loc_num+i]= list(data1[i])
            print(loc_num+i,"k값:",loc_num, "i값:",i,"//" "행에 지금 이 데이터 ==>",data1[i])
        loc_num+=len(data1)
        print("-------")
    # print(s[1])
```
